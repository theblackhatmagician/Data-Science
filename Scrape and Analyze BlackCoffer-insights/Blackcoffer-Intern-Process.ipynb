```python
#Remove Stop Words
```

```python
filtered_words = []
for i in data['token']:
    filtered_words.append([w.lower() for w in i if not w in total_stop and w.isalpha()])
```

```python
data['filtered_token'] = filtered_words 
```

```python
data.head()
```

```python
#Find Polarity, positive, negative and subjectivity
```

```python
positive = []
negative = []
polarity = []
subjectivity = []
words_count = []

for filtered_words in data['filtered_token']:
    pos = 0
    neg = 0
    words = 0
    for j in filtered_words:
        words += 1
        if j in positive_w:
            pos += 1
        elif j in negative_w:
            neg += 1
    polar = (pos-neg)/((pos+neg) + 0.000001)
    subject = (pos+neg)/(words + 0.000001)
    
    positive.append(pos)
    negative.append(neg)
    polarity.append(polar)
    subjectivity.append(subject)
    words_count.append(words)
```

```python
output['POSITIVE SCORE'] = positive
output['NEGATIVE SCORE'] = negative
output['POLARITY SCORE'] = polar
output['SUBJECTIVITY SCORE'] = subjectivity
output['WORD COUNT'] = words_count
output.head()
```

```python
#Count of Complex Words and Count of Syllable
```

```python
#pyphen package to find the syllable count
#Syllable count has certain rules than the count of vowels, Hence pyphen package is preferred
from pyphen import Pyphen
p = Pyphen(lang='en_US')
```

```python
complex_word = []
syllable_count = []
for doc in data['token']:
    complex_count = 0
    syllable = 0
    for i in doc:
        syllable += len(p.positions(i))+1
        if syllable > 2:
            complex_count+=1
    complex_word.append(complex_count)
    syllable_count.append(syllable)
```

```python
output['COMPLEX WORD COUNT'] = complex_word
#Total syllable count in each document is found, as it is not specified and a document has multiple words.
output['SYLLABLE PER WORD'] = syllable_count
output.head()
```

```python
#Sentence count and Word count for AVG NUMBER OF WORDS PER SENTENCE, FOG INDEX
```

```python
from nltk import sent_tokenize
```

```python
sentence_count = []
for i in data['Document']:
    sentences = sent_tokenize(i)
    sentence_count.append(len(sentences))
```

```python
word_length = []
char_length = []
for i in tokens:
    word_length.append(len(i))
    ch_count = 0
    for j in i:
        ch_count += len(j)
    char_length.append(ch_count)
```

```python
average_length = []
complex_percentage = []
fog_index = []
average_word_length = []
for i in range(150):
    average_length.append(word_length[i]/sentence_count[i])
    complex_percentage.append(complex_word[i]/word_length[i])
    fog_index.append(0.4*((word_length[i]/sentence_count[i])+(complex_word[i]/word_length[i])))
    average_word_length.append(char_length[i]/word_length[i])
```

```python
output['AVG NUMBER OF WORDS PER SENTENCE'] = average_length
output['AVG SENTENCE LENGTH'] = average_length
output['PERCENTAGE OF COMPLEX WORDS'] = complex_percentage
output['FOG INDEX'] = fog_index
output['AVG WORD LENGTH'] = average_word_length
output
```

```python
#Personal Pronouns
```

```python
pro_count = 0
for i in filtered_words:
    if i in ['i','we','my','ours','us']:
        pro_count += 1
pro_count
```

```python
import re
```

```python
pronoun_count = []
for i in data['Document']:
    #re.I makes case insensitive, but ?-i makes us case sensitive to avoid US
    pronounRegex = re.compile(r'\b(I|we|my|ours|(?-i:us))\b',re.I)
    pronoun = pronounRegex.findall(i)
    pronoun_count.append(len(pronoun))
```

```python
output['PERSONAL PRONOUNS'] = pronoun_count
```
