#### Handling Numeric Data

```python
#we can drop the car_ID variable which is not relevent for the price modeling
car_data.drop('car_ID',axis=1,inplace=True)
```

```python
#Creating avgmileage variable from citympg and highwaympg and dropping the two, to reduce the dimension.
car_data['averagemileage']=(car_data['citympg']+car_data['highwaympg'])/2
car_data = car_data.drop(['citympg','highwaympg'],axis='columns')
car_data.head()
```

```python
#HeatMap For Finding correlation of numeric type variables
plt.figure(figsize = (16, 10))
cor=car_data.corr()
sns.heatmap(cor, annot = True, cmap="YlGnBu")
plt.show()
```

From the Heatmap, highly correlated variables are identified.

```python
#Dropping carheight,stroke,peakrpm and compressionratio variable beacause they are not much correlated with price
car_data = car_data.drop(['stroke','peakrpm'],axis='columns')
```

#### Handling Categorical Data

```python
#Creating a generalized Variable car_company than CarName and dropping CarName
car_data['car_company'] = car_data['CarName'].apply(lambda x: x.split( )[0])
car_data.drop('CarName',axis=1,inplace=True)
```

```python
car_data['car_company'].unique()
```

Here Some Data are misspelled or abbrivated like vokswagen and vw

```python
# Correcting the misspelled company names.
car_data['car_company'].replace('maxda','mazda',inplace=True)
car_data['car_company'].replace('Nissan','nissan',inplace=True)
car_data['car_company'].replace('porcshce','porsche',inplace=True)
car_data['car_company'].replace('toyouta','toyota',inplace=True)
car_data['car_company'].replace(['vokswagen','vw'],'volkswagen',inplace=True)
```

```python
car_data['car_company'].unique()
```

```python
#Checking the correlation of Categorical variables with price
```

```python
#Creating dataframe with all categorical values
df_cat = car_data[list(car_data.columns[car_data.dtypes == 'object'])] 
df_cat[['symboling']] = car_data[['symboling']]
df_cat[['price']] = car_data[['price']]
```

```python
nominal.associations(df_cat,figsize=(30,10),mark_columns=True);
```

From The HeatMap We can identify the variables which has more correlation with the price. Other Variables can be dropped.
Good Correlated Variables:drivewheel,cylindernumber,fuelsystem,car_company

```python
#Dropping Bad Correlated Variables
```

```python
car_data = car_data.drop(['symboling','fueltype','aspiration','doornumber','enginelocation'],axis='columns')
```

```python
car_data
```

```python
#Converting categorical variables with integer levels
```

```python
# Categorizing car_company variable with an integer
car_labels = car_data['car_company'].astype('category').cat.categories.tolist()
replace_car_data = {'car_company' : {k: v for k,v in zip(car_labels,list(range(1,len(car_labels)+1)))}}

replace_car_data
```

```python
car_data.replace(replace_car_data, inplace = True)
car_data['car_company'] = car_data['car_company'].astype('category')
```

```python
car_data['drivewheel'].unique()
```

```python
car_data['drivewheel'] = car_data['drivewheel'].map({'rwd': 1, 'fwd': 2, '4wd':3})
car_data['drivewheel'] = car_data['drivewheel'].astype('category')
```

```python
car_data['enginetype'].unique()
```

```python
car_data['enginetype'] = car_data['enginetype'].map({'dohc':1, 'ohcv':2, 'ohc':3, 'l':4, 'rotor':5, 'ohcf':6, 'dohcv':7})
car_data['enginetype'] = car_data['enginetype'].astype('category')
```

```python
car_data['carbody'].unique()
```

```python
car_data['carbody'] = car_data['carbody'].map({'convertible':1, 'hatchback':2, 'sedan':3, 'wagon':4, 'hardtop':5})
car_data['carbody'] = car_data['carbody'].astype('category')
```

```python
car_data['cylindernumber'].unique()
```

```python
car_data['cylindernumber'] = car_data['cylindernumber'].map({'four':3, 'six':5, 'five':4, 'three':2, 'twelve':7, 'two':1, 'eight':6})
car_data['cylindernumber'] = car_data['cylindernumber'].astype('category')
```

```python
car_data['fuelsystem'].unique()
```

```python
car_data['fuelsystem'] = car_data['fuelsystem'].map({'mpfi':1, '2bbl':2, 'mfi':3, '1bbl':4, 'spfi':5, '4bbl':6, 'idi':7, 'spdi':8})
car_data['fuelsystem'] = car_data['fuelsystem'].astype('category')
```

```python
car_data.head()
```

### Data Splitting and Scaling

```python
from sklearn.preprocessing import MinMaxScaler
# Using MinMaxScaler to scale all the numeric variables in the same scale between 0 and 1.
scaler = MinMaxScaler()
```

```python
# Apply scaler() to all the columns except the categorical variables
num_vars = ['horsepower','wheelbase','curbweight', 'enginesize', 'boreratio','carlength','carwidth','averagemileage','price','carheight']
car_data[num_vars] = scaler.fit_transform(car_data[num_vars])
car_data.head()
```

```python
#Preparing independant variable
X= car_data.drop('price',axis='columns')
X.head()
```

```python
#Preparing dependant variable
Y= car_data.price
Y.head()
```

```python
#splitting training and test set
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=10)
```

### Making Regression Model

```python
#Model001
```

```python
import statsmodels.api as sm
```

```python
X_train_lm = sm.add_constant(X_train)
lr_1 = sm.OLS(Y_train, X_train_lm).fit()
```

```python
lr_1.summary()
```

```python
#R-squared value is 87%
#Analysing P value , there are varibales to be dropped
#Calculate VIF
#Check for the VIF values of the feature variables. 
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif = pd.DataFrame()
vif['Features'] = X_train.columns
vif['VIF'] = [variance_inflation_factor(X_train.values, i) for i in range(X_train.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```
